```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df1=pd.read_csv('../data/cleaned_data.csv')
```

```python
df2 = df1.sort_values(by=['player_id', 'fifa_version'])

columns_to_compare = [
    'weak_foot', 'skill_moves', 'pace', 'shooting', 'passing', 'dribbling', 'defending', 'physic',
    'attacking_crossing', 'attacking_finishing', 'attacking_heading_accuracy', 'attacking_short_passing',
    'attacking_volleys', 'skill_dribbling', 'skill_curve', 'skill_fk_accuracy', 'skill_long_passing',
    'skill_ball_control', 'movement_acceleration', 'movement_sprint_speed', 'movement_agility',
    'movement_reactions', 'movement_balance', 'power_shot_power', 'power_jumping', 'power_stamina',
    'power_strength', 'power_long_shots', 'mentality_aggression', 'mentality_interceptions',
    'mentality_positioning', 'mentality_vision', 'mentality_penalties', 'mentality_composure',
    'defending_marking_awareness', 'defending_standing_tackle', 'defending_sliding_tackle',
]


df2['improvement_1_year'] = 0
df2['improvement_2_years'] = 0
df2['improvement_3_years'] = 0


for player_id, group in df2.groupby('player_id'):
    rows = group[columns_to_compare].values  
    num_years = len(rows)
    
    for i in range(num_years):
        if i + 1 < num_years:
            # Calculate the improvements for 1 year
            improvement_1_year = rows[i + 1] - rows[i]
            df2.at[group.index[i], 'improvement_1_year'] = improvement_1_year.sum()
        
        if i + 2 < num_years:
            # Calculate the improvements for 2 years
            improvement_2_years = rows[i + 2] - rows[i]
            df2.at[group.index[i], 'improvement_2_years'] = improvement_2_years.sum()
        
        if i + 3 < num_years:
            # Calculate the improvements for 3 years
            improvement_3_years = rows[i + 3] - rows[i]
            df2.at[group.index[i], 'improvement_3_years'] = improvement_3_years.sum()
```
