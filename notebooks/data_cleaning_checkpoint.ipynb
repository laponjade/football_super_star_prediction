```python
# Task 1: Drop all samples where the position is a goalkeeper
print(f"Original dataset shape: {df.shape}")

# Filter out goalkeepers (drop rows where player_positions contains 'GK')
df_no_gk = df[~df['player_positions'].str.contains('GK', na=False)].copy()

print(f"Dataset shape after removing goalkeepers: {df_no_gk.shape}")
print(f"Number of goalkeepers removed: {len(df) - len(df_no_gk)}")

# Update df to the filtered version
df = df_no_gk
print(f"\nUpdated df shape: {df.shape}")
```

```python
# Task 2: Keep only necessary features for predicting if a player would become a super star or not

# Define the exact features to keep (as specified)
features_to_keep = [
    'player_id',
    'fifa_version',
    'player_positions',
    'overall',
    'potential',
    'age',
    'height_cm',
    'weight_kg',
    'preferred_foot',
    'weak_foot',
    'skill_moves',
    'pace',
    'shooting',
    'passing',
    'dribbling',
    'defending',
    'physic',
    'attacking_crossing',
    'attacking_finishing',
    'attacking_heading_accuracy',
    'attacking_short_passing',
    'attacking_volleys',
    'skill_dribbling',
    'skill_curve',
    'skill_fk_accuracy',
    'skill_long_passing',
    'skill_ball_control',
    'movement_acceleration',
    'movement_sprint_speed',
    'movement_agility',
    'movement_reactions',
    'movement_balance',
    'power_shot_power',
    'power_jumping',
    'power_stamina',
    'power_strength',
    'power_long_shots',
    'mentality_aggression',
    'mentality_interceptions',
    'mentality_positioning',
    'mentality_vision',
    'mentality_penalties',
    'mentality_composure',
    'defending_marking_awareness',
    'defending_standing_tackle',
    'defending_sliding_tackle',
    
    # Note: 'concatenated' feature will be added later if needed
]

# Check which features exist in the current dataframe
existing_features_to_keep = [f for f in features_to_keep if f in df.columns]
missing_features = [f for f in features_to_keep if f not in df.columns]

print(f"Features to keep: {len(existing_features_to_keep)}")
if missing_features:
    print(f"Warning: {len(missing_features)} features not found in dataset: {missing_features}")

# Keep only the specified features
df_cleaned = df[existing_features_to_keep].copy()

print(f"\nDataset shape before feature selection: {df.shape}")
print(f"Dataset shape after feature selection: {df_cleaned.shape}")
print(f"Number of features removed: {df.shape[1] - df_cleaned.shape[1]}")

# Update df to the cleaned version
df = df_cleaned
print(f"\nUpdated df shape: {df.shape}")
print(f"\nRemaining columns ({len(df.columns)}):")
for i, col in enumerate(df.columns, 1):
    print(f"{i:3d}. {col}")
```

```python
# Task 3: Apply one-hot encoding technique for 3 features: attacker_position, midfielder_position, defender_position
# Highly optimized version using fully vectorized string operations

# Define position categories
attacker_positions = ['ST', 'CF', 'LW', 'RW', 'LF', 'RF', 'LS', 'RS']
midfielder_positions = ['CAM', 'CM', 'CDM', 'LM', 'RM', 'LAM', 'RAM', 'LCM', 'RCM', 'LDM', 'RDM']
defender_positions = ['CB', 'LB', 'RB', 'LCB', 'RCB', 'LWB', 'RWB']

# Fill NaN values with empty string for vectorized operations
positions_str = df['player_positions'].fillna('')

# Use vectorized string operations to check if position strings contain any category positions
# This is much faster than iterating through rows
df['attacker_position'] = (
    positions_str.str.contains('|'.join(attacker_positions), case=False, na=False, regex=True)
).astype(int)

df['midfielder_position'] = (
    positions_str.str.contains('|'.join(midfielder_positions), case=False, na=False, regex=True)
).astype(int)

df['defender_position'] = (
    positions_str.str.contains('|'.join(defender_positions), case=False, na=False, regex=True)
).astype(int)

# Display results
print("One-hot encoding applied successfully (highly optimized version)!")
print(f"\nPosition distribution:")
print(f"Attacker positions: {df['attacker_position'].sum()} players")
print(f"Midfielder positions: {df['midfielder_position'].sum()} players")
print(f"Defender positions: {df['defender_position'].sum()} players")

# Show some examples
print(f"\nSample of position encoding:")
sample_cols = ['player_positions', 'attacker_position', 'midfielder_position', 'defender_position']
print(df[sample_cols].head(10))

print(f"\nFinal dataset shape: {df.shape}")
```

```python
# Apply ordinal encoding to preferred_foot feature using scikit-learn's OrdinalEncoder
from sklearn.preprocessing import OrdinalEncoder

# Check unique values in preferred_foot before encoding
print("Unique values in preferred_foot before encoding:")
print(df['preferred_foot'].value_counts())
print(f"\nData type: {df['preferred_foot'].dtype}")

# Initialize OrdinalEncoder
ordinal_encoder = OrdinalEncoder()

# Reshape the column to 2D array (required by sklearn)
preferred_foot_2d = df[['preferred_foot']].values

# Fit and transform the data
preferred_foot_encoded = ordinal_encoder.fit_transform(preferred_foot_2d)

# Replace the original column with encoded values
df['preferred_foot'] = preferred_foot_encoded.astype(int)

# Display the mapping (categories in order)
print(f"\nOrdinal encoding mapping:")
categories = ordinal_encoder.categories_[0]
for idx, category in enumerate(categories):
    print(f"  {category} -> {idx}")

# Display results
print(f"\nOrdinal encoding applied successfully using scikit-learn!")
print(f"\nPreferred foot distribution after encoding:")
print(df['preferred_foot'].value_counts().sort_index())
print(f"\nUpdated dataset shape: {df.shape}")
```

```python
# Check for duplicate rows for a single player within each FIFA version

# Group by fifa_version and player_id and count occurrences
duplicate_check = df.groupby(['fifa_version', 'player_id']).size().reset_index(name='count')

# Find duplicates (count > 1)
duplicates = duplicate_check[duplicate_check['count'] > 1]

print(f"Total unique (fifa_version, player_id) combinations: {len(duplicate_check)}")
print(f"Number of duplicate (fifa_version, player_id) combinations: {len(duplicates)}")

if len(duplicates) > 0:
    print(f"\n⚠️ Found {len(duplicates)} duplicate player entries within FIFA versions!")
    print("\nDuplicate entries:")
    print(duplicates.head(20))
    
    # Show the actual duplicate rows
    print("\n" + "="*80)
    print("Sample of actual duplicate rows:")
    print("="*80)
    
    # Get first few duplicate cases
    sample_duplicates = duplicates.head(5)
    for idx, row in sample_duplicates.iterrows():
        fifa_ver = row['fifa_version']
        player_id = row['player_id']
        print(f"\nFIFA Version: {fifa_ver}, Player ID: {player_id} ({row['count']} occurrences):")
        duplicate_rows = df[(df['fifa_version'] == fifa_ver) & (df['player_id'] == player_id)]
        print(duplicate_rows[['player_id', 'fifa_version', 'overall', 'potential', 'age']].head())
else:
    print("\n✅ No duplicate player entries found within FIFA versions!")
    print("Each player appears only once per FIFA version.")
```

```python
# Delete all samples where fifa_version is 15 or 16

print(f"Dataset shape before filtering: {df.shape}")
print(f"\nFIFA version distribution before filtering:")
print(df['fifa_version'].value_counts().sort_index())

# Count rows to be removed
rows_to_remove = df[df['fifa_version'].isin([15.0, 16.0])]
print(f"\nRows to be removed (fifa_version 15 or 16): {len(rows_to_remove)}")

# Filter out fifa_version 15 and 16
df_filtered = df[~df['fifa_version'].isin([15.0, 16.0])].copy()

print(f"\nDataset shape after filtering: {df_filtered.shape}")
print(f"Rows removed: {len(df) - len(df_filtered)}")

# Update df
df = df_filtered

print(f"\nFIFA version distribution after filtering:")
print(df['fifa_version'].value_counts().sort_index())
print(f"\nUpdated df shape: {df.shape}")
```
